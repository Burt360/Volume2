```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(lam):
    N = 100
    x = np.linspace(0, 10, 500)
    y = np.abs(np.sin(np.pi*(lam-x)) / (N*np.sin(np.pi*(lam-x)/N)) )

    plt.plot(x, y)
    plt.xlabel('k')
    plt.ylabel('| f(k) |')
    plt.title(f'lambda = {lam}')
    plt.tight_layout()
    plt.show()

for lam in (3/2, 7/4, 2.2334931):
    plot(lam)
```
