```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

N = (1,2,4,8,16,32)
```

```python
### Uniform([0,1])
a, b = 0, 1

# mean, variance
mu = (a+b)/2
s2 = (b-a)**2 / 12

# Create subplots
fig, axs = plt.subplots(2,3)

# x points
x = np.linspace(0, 1, 100)

# PDF of Uniform([0,1])
y = stats.uniform(a, b).pdf(x)

for ax, n in zip(axs.flatten(), N):
    # Plot PDF of Uniform([0, 1])
    ax.plot(x, y)

    # Plot PDF of Normal(mu, s2/n**(1/2)) 
    y_normal = stats.norm(mu, s2/n**(1/2)).pdf(x)
    ax.plot(x, y_normal)

    # Compute the samples of size n and their means
    samples = [np.random.uniform(a, b, n) for _ in range(1000)]
    sample_means = [np.mean(sample) for sample in samples]

    # Create a histogram of the sample means
    ax.hist(sample_means, density=True)

    ax.set_title(f'n = {n}')
    
fig.set_size_inches(15,10)
plt.show()
```

```python
### Beta(1/2, 1/2)
a, b = 1/2, 1/2

# mean, variance
mu = a/(a+b)
s2 = a*b/((a+b)**2 * (a+b+1))

# Create subplots
fig, axs = plt.subplots(2,3)

# x points
x = np.linspace(0, 1, 1000)

# PDF of Beta(1/2, 1/2)
y = stats.beta(a, b).pdf(x)

for ax, n in zip(axs.flatten(), N):
    # Plot PDF of Beta(1/2, 1/2)
    ax.plot(x, y)

    # Plot PDF of Normal(mu, s2/n**(1/2)) 
    y_normal = stats.norm(mu, s2/n**(1/2)).pdf(x)
    ax.plot(x, y_normal)

    # Compute the samples of size n and their means
    samples = [np.random.beta(a, b, n) for _ in range(1000)]
    sample_means = [np.mean(sample) for sample in samples]

    # Create a histogram of the sample means
    ax.hist(sample_means, density=True)

    ax.set_title(f'n = {n}')
    
fig.set_size_inches(15,10)
plt.show()
```
